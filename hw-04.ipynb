**Завдання 1. Аналітика помилок у логах.** \
Маємо загальну кількість подій за добу $n(Ω)=500$. \
З них:
*   помилки тільки бази даних з тегом $DB$: 85
*   помилки  тільки мережи з тегом $NET$: 60
*   помилки з обома тегами (перетин $A\cap B$ ): 35
- - -
**1. Яка ймовірність, що випадково вибрана подія - помилка $DB$?** \
Подія $A$ включає як тільки $DB$, так і ті що мають обидва теги. Кількість подій А: $$n(A)=85+35=120$$ \
Ймовірність $P(A)=\frac{n(A)}{n(Ω)}=\frac{120}{500}=0.24\cdot 100\%=24\%$

**2. Яка ймовірність, що подія має тег $DB$ або $NET$?** \
Це об’єднання множин  $A\cup B$. Використовуємо формулу:
$$P(A \cup B)=\frac {|A|+|B|-|A \cap B|}{500}$$
де: \
$|A|=120, |B|=60+35=95, |A \cap B|=35.$

$$P(A \cup B)=\frac {(120+95-35)}{500}=\frac {180}{500}=0.36 ⋅100 \%=36\% $$

**3. Ймовірність події з тегом $DB$, але не $NET$?** \
Це різниця множин $A/B$, тобто за умовами завдання це помилки тільки бази даних з тегом  $DB$:
$$ P(A/B)=\frac {85}{500}=0.17 \cdot 100\%=17\% $$

**Відповідь:** \
Помилка $DB (A) = 24\%$; \
$DB$ або $NET$ $(A \cup B) = 36\%$; \
$DB$ без $NET$ $(A/B) = 17\%$.

**Завдання 2. Контроль якості батарей** \
Умова: \
**Маємо 12 батарей: 8 справних $(S)$ і 4 дефектних $(D)$. Для тестування випадково вибирають 3 батареї без повернення.**
- - -

Знайдемо ймовірність:
1. Що всі 3 батареї $S$ справні (через умовні ймовірності):
$$ P(S_1 ∩ S_2 \cap S_3)=\frac{8}{12}\cdot \frac{7}{11} \cdot \frac{6}{10}=\frac {14}{55} \approx 0.2545 \cdot 100\%=25.45 \% $$

2. Комбінаторний метод $(C_n^k)$. \
Загальна кількість способів вибрати 3 батареї з 12:
$$ C_8^3=\frac {12}{3} \cdot \frac {11}{2} \cdot \frac {10}{1}=220$$

$$ C_8^3=\frac {8}{3} \cdot \frac {7}{2} \cdot \frac {6}{1}=56$$
Ймовірність:
$$P=\frac{56}{220}=\frac{15}{55} ≈ 0.2545 \cdot 100\%=25.45 \% $$

**Висновок:** Результати обох методів ідентичні, що підтверждує правильність розрахунків.

$$P(\bar{A_3})=0.97$$
$$P(\bar{A_4})=0.96$$

$$ P(всі\ працюють)=P(\bar{A_1})\cdot P(\bar{A_2}) \cdot P(\bar{A_3})\cdot P(\bar{A_4}) $$

$$ P(хоча\ б\ один\ відмова)=1-P(всі\ працюють) $$
$$ P(хоча\ б\ один\ відмова)=1-0,8669=0,1331 $$

**Завдання 4. Визначення джерела трафіку.**
Маємо:
*   Пошукова реклама: 50% трафіку;
*   Соцмережі:30% трафіку;
*   Email-розсилка:20% трафіку.

Конверсія (ймовірність покупки) для кожного джерела:

*  Пошукова реклама: 4%;
*  Соцмережі:2%;
*  Email-розсилка: 8%.

Позначимо:
*  $H_1$ - подія "відвідувач прийшов з пошукової реклами";
*  $H_2$ - подія "відвідувач прийшов з соцмереж";
*  $H_3$ - подія "відвідувач прийшов з email-розсилки";
*  $A$ - подія "відвідувач зробив покупку".

- - -
**1. Ймовірність, що випадковий відвідувач зробить покупку.** \
Для обчислення загальної ймовірності використаємо формулу повної ймовірності:
$$P(A)=\sum_{i=1}^{n}P(H_i)⋅P(A|H_i) $$
де:
*   $P(H_1)=0.5 (пошукова\ реклама),\ P(A|H_1)= 0,04 $

*   $P(H_2)=0.3 (соцмережі),\ P(A|H_2)= 0,02 $

*   $P(H_3)=0.2 (email-розсилка),\  P(A|H_3)= 0,08 $

Розрахунок:
$$ P(A)=(0.5\cdot0.04)+(0.3⋅0.02)+(0.2 ⋅0.08)$$
$$ P(A)=0.02+0.006+0.016=0.042⋅100\%=4.2\%$$

**2. Ймовірність, що відвідувач, який зробить покупку прийшов з email-розсилки.** \
Знайдемо ймовірність за теоремою Байєса: \
$$ P(H_3|A)=\frac{P(H_3)⋅P(A|H_3)}{P(A)} $$
$$ P(H_3|A)=\frac{0.2\cdot0.08}{0.042}= \frac {0.016}{0.042}≈0.381\cdot100\%≈38.1\%$$


```python
# Завдання №3. Функція, яка для довільної кількості джерел обчислює апостеріорні ймовірності кожного джерела після здійснення покупки.
def bayes_source(prior_probs, conversion_rates):
    if len(prior_probs) != len(conversion_rates):
        raise ValueError("prior_probs і conversion_rates мають бути однакової довжини")
    if len(prior_probs) == 0:
        return []

    for p in prior_probs:
        if p < 0:
            raise ValueError("У prior_probs не може бути від'ємних значень")
    for c in conversion_rates:
        if not (0 <= c <= 1):
            raise ValueError("conversion_rates мають бути в діапазоні [0, 1]")

    # P(A) за формулою повної ймовірності
    p_buy = sum(p * c for p, c in zip(prior_probs, conversion_rates))
    if p_buy == 0:
        raise ValueError("P(A)=0: неможливо обчислити апостеріорні ймовірності")

    # P(H_i | A) за Байєсом
    posteriors = [(p * c) / p_buy for p, c in zip(prior_probs, conversion_rates)]
    return posteriors
print(bayes_source([0.5, 0.3, 0.2], [0.04, 0.02, 0.08]))
```
