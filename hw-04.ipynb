**Завдання 1. Аналітика помилок у логах.** \
Маємо загальну кількість подій за добу $n(Ω)=500$. \
З них:
*   помилки тільки бази даних з тегом $DB$: 85
*   помилки  тільки мережи з тегом $NET$: 60
*   помилки з обома тегами (перетин $A⋂B$ ): 35
- - -
**1. Яка ймовірність, що випадково вибрана подія - помилка $DB$?** \
Подія $A$ включає як тільки $DB$, так і ті що мають обидва теги. Кількість подій А: $$n(A)=85+35=120$$ \
Ймовірність $P(A)=\frac{n(A)}{n(Ω)}=\frac{120}{500}=0.24\cdot 100\%=24\%$

**2. Яка ймовірність, що подія має тег $DB$ або $NET$?** \
Це об’єднання множин  $A\cup B$. Використовуємо формулу:
$$P(A \cup B)=\frac {|A|+|B|-|A \cap B|}{500}$$
де: \
$|A|=120, |B|=60+35=95, |A \cap B|=35.$

$$P(A \cup B)=\frac {(120+95-35)}{500}=\frac {180}{500}=0.36 ⋅100 \%=36\% $$

**3. Ймовірність події з тегом $DB$, але не $NET$?** \
Це різниця множин $A/B$, тобто за умовами завдання це помилки тільки бази даних з тегом  $DB$:
$$ P(A/B)=\frac {85}{500}=0.17 \cdot 100\%=17\% $$

**Відповідь:** \
Помилка $DB (A) = 24\%$; \
$DB$ або $NET$ $(A \bigcup B) = 36\%$; \
$DB$ без $NET$ $(A/B) = 17\%$.

**Завдання 2. Контроль якості батарей** \
Умова: \
**Маємо 12 батарей: 8 справних $(S)$ і 4 дефектних $(D)$. Для тестування випадково вибирають 3 батареї без повернення.**
- - -

Знайдемо ймовірність:
1. Що всі 3 батареї $S$ справні (через умовні ймовірності):
$$ P(S_1 ∩ S_2 \cap S_3)=\frac{8}{12}\cdot \frac{7}{11} \cdot \frac{6}{10}=\frac {14}{55} \approx 0.2545 \cdot 100\%=25.45 \% $$

2. Комбінаторний метод $(C_n^k)$. \
Загальна кількість способів вибрати 3 батареї з 12:
$$ C_8^3=\frac {12}{3} \cdot \frac {11}{2} \cdot \frac {10}{1}=220$$
Сприятливі випадки (вибрати 3 батареї з 8 справних):
$$ C_8^3=\frac {8}{3} \cdot \frac {7}{2} \cdot \frac {6}{1}=56$$
Ймовірність:
$$P=\frac{56}{220}=\frac{15}{55} ≈ 0.2545 \cdot 100\%=25.45 \% $$

**Висновок:** Результати обох методів ідентичні, що підтверждує правильність розрахунків.

**Яка ймовірність, що рівно 2 батареї справні?** \
3. У наборі з 3-х батарей має бути 2 справні $(S)$ та 1 дефектна $(D)$
Можливі комбінації порядку та ймовірність появи дефектної батареї:

**Спосіб 1. Правило множення**

$$ P (S, S, D)= \frac {8}{12} \cdot \frac {7}{11} \cdot \frac {4}{10} =\frac {224}{1320} $$

$$ P (S, D, S)= \frac {8}{12} \cdot \frac {4}{11} \cdot \frac {7}{10} =\frac {224}{1320} $$

$$ P (D, S, S)= \frac {4}{12} \cdot \frac {8}{11} \cdot \frac {7}{10} =\frac {224}{1320} $$

$$ P = \frac {224}{1320} + \frac {224}{1320} + \frac {224}{1320} =\frac {672}{1320}=\frac {28}{55} \approx0.5091 \cdot 100\% \approx 50.91\% $$

**Спосіб 2. Комбінаторний метод**

Вибираємо 2 справін батареї з 8 та 1 дефектну з 4.

Кількість способів:
$$ C_8^2 \cdot C^1_4=\frac {8 \cdot 7}{2}\cdot 4=28\cdot 4 =112$$

Ймовірність:
$$P=\frac {C_8^2 \cdot C_4^1}{C_8^3} $$
$$ C_8^3=\frac {3}{12} \cdot \frac{2}{11} \cdot \frac{1}{10}= \frac {6}{1320}=220 $$
$$P=\frac {112}{220}=\approx0.5091 \cdot 100\% \approx 50.91\% $$

**Завдання 3. Надійність розподіленої системи** \
Маємо ймовірності відмови кожного вузла протягом однієї години: \
Вузол А: $P(A_1)=0.02$ \
Вузол B: $P(A_2)=0.05$ \
Вузол C: $P(A_3)=0.03$ \
Вузол D: $P(A_4)=0.04$
- - -
**1. Ймовірність, щро протягом години всі 4 вузли працюють.** \
Ймовірність безперевної роботи - це протилежна подія до відмови протягом однієї години $P(\bar{A_i})=1-P(A_i)$:
$$P(\bar{A_1})=0.98$$
$$P(\bar{A_2})=0.95$$
$$P(\bar{A_3})=0.97$$
$$P(\bar{A_4})=0.96$$

$$ P(всі\ працюють)=P(\bar{A_1})\cdot P(\bar{A_2}) \cdot P(\bar{A_3})\cdot P(\bar{A_4}) $$
$$ P(всі\ працюють)=0.98 \cdot 0.95 \cdot 0.97 \cdot 0.96 \approx 0.8669 \cdot 100\% \approx 86.69 \%$$

**2. Ймовірність, що протягом години хоча б один вузол відмовить**
Подія, при який хоча б один вузол відмовить" протилежна до події коли всі вузли працюють, тому:
$$ P(хоча\ б\ один\ відмова)=1-P(всі\ працюють) $$
$$ P(хоча\ б\ один\ відмова)=1-0,8669=0,1331 $$

**3. Ймовірність, що протягом години відмовлять рівно два вузли?** \
Потрібно врахувати всі можливі комбінації, де два вузла відмовили $(q)$, а два працюють $(p)$:
*   A, B відмовили: $q_1, q_2, p_3, p_4 =0.02 ⋅0.05 \cdot 0.97 \cdot 0.96 =0.0009312 $
*   A, C відмовили: $q_1, p_2, q_3, p_4 =0.02 ⋅0.95 \cdot 0.03 \cdot 0.96 =0.0005472 $
*   A, D відмовили: $q_1, p_2, p_3, q_4 =0.02 ⋅0.95 \cdot 0.97 \cdot 0.04 =0.0007372 $
*   B, C відмовили: $p_1, q_2, q_3, p_4 =0.98 ⋅0.05 \cdot 0.03 \cdot 0.96 =0.0014112 $
*   B, D відмовили: $p_1, q_2, p_3, q_4 =0.98 ⋅0.05 \cdot 0.97 \cdot 0.04 =0.0019012 $
*   C, D відмовили: $p_1, p_2, q_3, q_4 =0.98 ⋅0.95 \cdot 0.03 \cdot 0.04 =0.0011172 $

Додамо всі результати:
$$ P(рівно\ два)=0.0009312+0.0005472+0.0007372+0.0014112+0.0019012+0.0011172=0.0066452$$
$$ P(рівно\ два)=0.00066452 \cdot 100\% \approx 0.66\%  $$
